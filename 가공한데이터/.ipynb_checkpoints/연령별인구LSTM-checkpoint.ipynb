## 동별로 시계열 맞춰야하니까 각 동별로 나누고, 연령코드로 나누고 맞춘값들 평균내서 보자

```python
def prepro(df, dong, age):
    d1 = df[df.행정동명 == dong]
    d2 = d1[d1.연령코드 == age]
    d3 = d2[['Datetime','총인구수']]
    d3.set_index('Datetime',inplace=True)
    return d3
```

```python
test3 = prepro(data_df, '독산제1동', 15)
test3
```

```python
plt.figure(figsize=[100, 20])
plt.plot(np.arange(0, 98), test3)
plt.show()
```

```python
split_date = pd.Timestamp('01-06-2018')

train = test3.loc[:split_date, ['총인구수']]
test = test3.loc[split_date:, ['총인구수']]

ax = train.plot()
test.plot(ax=ax)
plt.legend(['train', 'test'])
```

```python
sc = MinMaxScaler()

train_sc = sc.fit_transform(train)
test_sc = sc.transform(test)

train_sc
```

```python
train_sc_df = pd.DataFrame(train_sc, columns=['Scaled'], index=train.index)
test_sc_df = pd.DataFrame(test_sc, columns=['Scaled'], index=test.index)
train_sc_df.head()
```

```python
for s in range(1, 13):
    train_sc_df['shift_{}'.format(s)] = train_sc_df['Scaled'].shift(s)
    test_sc_df['shift_{}'.format(s)] = test_sc_df['Scaled'].shift(s)

train_sc_df.head(13)
```

```python
X_train = train_sc_df.dropna().drop('Scaled', axis=1)
y_train = train_sc_df.dropna()[['Scaled']]

X_test = test_sc_df.dropna().drop('Scaled', axis=1)
y_test = test_sc_df.dropna()[['Scaled']]
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train = np.array(X_train)
X_test= np.array(X_test)

y_train = np.array(y_train)
y_test = np.array(y_test)
print(X_train.shape)
print(X_train)
print(y_train.shape)
print(y_train)
```

```python
X_train_t = X_train.reshape(X_train.shape[0], 12, 1) 
X_test_t = X_test.reshape(X_test.shape[0], 12, 1)
```

```python
print("최종 DATA")
print(X_train_t.shape)
print(X_train_t)
print(y_train)
```

```python
K.clear_session() 
model = Sequential()
model.add(LSTM(30, input_shape=(12, 1))) 
# model.add(Dropout(0.3))
model.add(Dense(12))
model.add(Dense(1)) 
model.compile(loss='mean_squared_error', optimizer='adam') 
model.summary()
```

```python
early_stop = EarlyStopping(monitor='loss', patience=3, verbose=1)

model.fit(X_train_t, y_train, epochs=100,
          batch_size=1, verbose=1, callbacks=[early_stop])
```

```python
y_hat = model.predict(X_test_t)
y_axis = np.arange(0, len(y_train))
y_2_axis = np.arange(len(y_train), len(y_train) + len(y_hat))

plt.figure(figsize=(100,10))
plt.plot(y_axis, y_train.reshape(len(X_train_t),), '-o') #Train데이터 그래프로 표현
plt.plot(y_2_axis, y_hat.reshape(len(X_test_t),), '-o', color='red', label='predicted') #x_test 예측값 그래프로 표현
plt.plot(y_2_axis, y_test.reshape(len(X_test_t),), '-o', color='purple', alpha=0.3, label='actual') #y_test 실제값 그래프로 표현 
plt.legend(fontsize=40)
plt.show()
```

# 함수로 만드려는 시도중...

```python
def train_test_LSTM_model(df, timestamp='01-06-2018', month = 13):
    split_date = pd.Timestamp(timestamp)

    train = df.loc[:split_date, ['총인구수']]
    test = df.loc[split_date:, ['총인구수']]
    
    sc = MinMaxScaler()
    train_sc = sc.fit_transform(train)
    test_sc = sc.transform(test)
    
    train_sc_df = pd.DataFrame(train_sc, columns=['Scaled'], index=train.index)
    test_sc_df = pd.DataFrame(test_sc, columns=['Scaled'], index=test.index)
    
    for s in range(1, month):
        train_sc_df['shift_{}'.format(s)] = train_sc_df['Scaled'].shift(s)
        test_sc_df['shift_{}'.format(s)] = test_sc_df['Scaled'].shift(s)
    
    X_train = train_sc_df.dropna().drop('Scaled', axis=1)
    y_train = train_sc_df.dropna()[['Scaled']]
    X_test = test_sc_df.dropna().drop('Scaled', axis=1)
    y_test = test_sc_df.dropna()[['Scaled']]
    
    X_train = np.array(X_train)
    X_test= np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    X_train_t = X_train.reshape(X_train.shape[0], month-1, 1) 
    X_test_t = X_test.reshape(X_test.shape[0], month-1, 1)
    
    K.clear_session() 
    model = Sequential()
    model.add(LSTM(12, input_shape=(12, 1))) 
    # model.add(Dropout(0.3))
    model.add(Dense(5))
    model.add(Dense(1)) 
    model.compile(loss='mean_squared_error', optimizer='adam') 
    return X_train_t, X_test_t, y_train, y_test, model
```
