# 함수로 만들기

```python
def LSTM_model(df, dong, timestamp='01-06-2018', month = 13):
    d1 = df[df.행정동명 == dong]
    temp_df = d1[['Datetime','총인구수']]
    temp_df.set_index('Datetime',inplace=True)
    
    split_date = pd.Timestamp(timestamp)

    train = temp_df.loc[:split_date, ['총인구수']]
    test = temp_df.loc[split_date:, ['총인구수']]
    
    sc = MinMaxScaler()
    train_sc = sc.fit_transform(train)
    test_sc = sc.transform(test)
    
    train_sc_df = pd.DataFrame(train_sc, columns=['Scaled'], index=train.index)
    test_sc_df = pd.DataFrame(test_sc, columns=['Scaled'], index=test.index)
    
    for s in range(1, month):
        train_sc_df['shift_{}'.format(s)] = train_sc_df['Scaled'].shift(s)
        test_sc_df['shift_{}'.format(s)] = test_sc_df['Scaled'].shift(s)
    
    X_train = train_sc_df.dropna().drop('Scaled', axis=1)
    y_train = train_sc_df.dropna()[['Scaled']]
    X_test = test_sc_df.dropna().drop('Scaled', axis=1)
    y_test = test_sc_df.dropna()[['Scaled']]
    
    X_train = np.array(X_train)
    X_test= np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    X_train_t = X_train.reshape(X_train.shape[0], month-1, 1) 
    X_test_t = X_test.reshape(X_test.shape[0], month-1, 1)
    
    K.clear_session() 
    model = Sequential()
    model.add(LSTM(30, input_shape=(month-1, 1))) 
    # model.add(Dropout(0.3))
    model.add(Dense(12))
    model.add(Dense(1)) 
    model.compile(loss='mean_squared_error', optimizer='adam') 
    
    early_stop = EarlyStopping(monitor='loss', patience=10, verbose=1)
    model.fit(X_train_t, y_train, epochs=100,
              batch_size=1, verbose=1, callbacks=[early_stop])
    y_hat = model.predict(X_test_t)
    y_axis = np.arange(0, len(y_train))
    y_2_axis = np.arange(len(y_train), len(y_train) + len(y_hat))

    plt.figure(figsize=(30,10))
    plt.title(label='19, 20년도 {}의 5-14세 인구 예측결과'.format(dong), fontdict={'fontsize':50})
    plt.plot(y_axis, y_train.reshape(len(X_train_t),), '-o', label='학습데이터') #Train데이터 그래프로 표현
    plt.plot(y_2_axis, y_hat.reshape(len(X_test_t),), '-o', color='red', label='예측인구') #x_test 예측값 그래프로 표현
    plt.plot(y_2_axis, y_test.reshape(len(X_test_t),), '-o', color='purple', alpha=0.3, label='실제인구') #y_test 실제값 그래프로 표현 
    plt.legend(fontsize=40)
    plt.show()
    
```

```python
LSTM_model(data_need, '삼각산동', timestamp='01-09-2018', month = 4)
```

```python
def GRU_model(df, dong, timestamp='01-06-2018', month = 13):
    d1 = df[df.행정동명 == dong]
    temp_df = d1[['Datetime','총인구수']]
    temp_df.set_index('Datetime',inplace=True)
    
    split_date = pd.Timestamp(timestamp)

    train = temp_df.loc[:split_date, ['총인구수']]
    test = temp_df.loc[split_date:, ['총인구수']]
    
    sc = MinMaxScaler()
    train_sc = sc.fit_transform(train)
    test_sc = sc.transform(test)
    
    train_sc_df = pd.DataFrame(train_sc, columns=['Scaled'], index=train.index)
    test_sc_df = pd.DataFrame(test_sc, columns=['Scaled'], index=test.index)
    
    for s in range(1, month):
        train_sc_df['shift_{}'.format(s)] = train_sc_df['Scaled'].shift(s)
        test_sc_df['shift_{}'.format(s)] = test_sc_df['Scaled'].shift(s)
    
    X_train = train_sc_df.dropna().drop('Scaled', axis=1)
    y_train = train_sc_df.dropna()[['Scaled']]
    X_test = test_sc_df.dropna().drop('Scaled', axis=1)
    y_test = test_sc_df.dropna()[['Scaled']]
    
    X_train = np.array(X_train)
    X_test= np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    X_train_t = X_train.reshape(X_train.shape[0], month-1, 1) 
    X_test_t = X_test.reshape(X_test.shape[0], month-1, 1)
    
    K.clear_session() 
    model = Sequential()
    model.add(GRU(30, input_shape=(month-1, 1))) 
    # model.add(Dropout(0.3))
    model.add(Dense(12))
    model.add(Dense(1)) 
    model.compile(loss='mean_squared_error', optimizer='adam') 
    
    early_stop = EarlyStopping(monitor='loss', patience=10, verbose=1)
    model.fit(X_train_t, y_train, epochs=100,
              batch_size=1, verbose=1, callbacks=[early_stop])
    y_hat = model.predict(X_test_t)
    y_axis = np.arange(0, len(y_train))
    y_2_axis = np.arange(len(y_train), len(y_train) + len(y_hat))

    plt.figure(figsize=(30,10))
    plt.title(label='19, 20년도 {}의 5-14세 인구 예측결과'.format(dong), fontdict={'fontsize':50})
    plt.plot(y_axis, y_train.reshape(len(X_train_t),), '-o', label='학습데이터') #Train데이터 그래프로 표현
    plt.plot(y_2_axis, y_hat.reshape(len(X_test_t),), '-o', color='red', label='예측인구') #x_test 예측값 그래프로 표현
    plt.plot(y_2_axis, y_test.reshape(len(X_test_t),), '-o', color='purple', alpha=0.3, label='실제인구') #y_test 실제값 그래프로 표현 
    plt.legend(fontsize=40)
    plt.show()
    
```

```python
def Bi_LSTM_model(df, dong, timestamp='01-06-2018', month = 13):
    d1 = df[df.행정동명 == dong]
    temp_df = d1[['Datetime','총인구수']]
    temp_df.set_index('Datetime',inplace=True)
    
    split_date = pd.Timestamp(timestamp)

    train = temp_df.loc[:split_date, ['총인구수']]
    test = temp_df.loc[split_date:, ['총인구수']]
    
    sc = MinMaxScaler()
    train_sc = sc.fit_transform(train)
    test_sc = sc.transform(test)
    
    train_sc_df = pd.DataFrame(train_sc, columns=['Scaled'], index=train.index)
    test_sc_df = pd.DataFrame(test_sc, columns=['Scaled'], index=test.index)
    
    for s in range(1, month):
        train_sc_df['shift_{}'.format(s)] = train_sc_df['Scaled'].shift(s)
        test_sc_df['shift_{}'.format(s)] = test_sc_df['Scaled'].shift(s)
    
    X_train = train_sc_df.dropna().drop('Scaled', axis=1)
    y_train = train_sc_df.dropna()[['Scaled']]
    X_test = test_sc_df.dropna().drop('Scaled', axis=1)
    y_test = test_sc_df.dropna()[['Scaled']]
    
    X_train = np.array(X_train)
    X_test= np.array(X_test)
    y_train = np.array(y_train)
    y_test = np.array(y_test)
    
    X_train_t = X_train.reshape(X_train.shape[0], month-1, 1) 
    X_test_t = X_test.reshape(X_test.shape[0], month-1, 1)
    
    K.clear_session() 
    model = Sequential()
    model.add(Bidirectional(LSTM(30, return_sequences=True),input_shape=(month-1, 1)))
    # model.add(Dropout(0.3))
    model.add(Dense(12))
    model.add(Dense(1)) 
    model.compile(loss='mean_squared_error', optimizer='adam') 
    
    early_stop = EarlyStopping(monitor='loss', patience=10, verbose=1)
    model.fit(X_train_t, y_train, epochs=100,
              batch_size=1, verbose=1, callbacks=[early_stop])
    y_hat = model.predict(X_test_t)
    y_axis = np.arange(0, len(y_train))
    y_2_axis = np.arange(len(y_train), len(y_train) + len(y_hat))

    plt.figure(figsize=(30,10))
    plt.title(label='19, 20년도 {}의 5-14세 인구 예측결과'.format(dong), fontdict={'fontsize':50})
    plt.plot(y_axis, y_train.reshape(len(X_train_t),), '-o', label='학습데이터') #Train데이터 그래프로 표현
    plt.plot(y_2_axis, y_hat.reshape(len(X_test_t),), '-o', color='red', label='예측인구') #x_test 예측값 그래프로 표현
    plt.plot(y_2_axis, y_test.reshape(len(X_test_t),), '-o', color='purple', alpha=0.3, label='실제인구') #y_test 실제값 그래프로 표현 
    plt.legend(fontsize=40)
    plt.show()
    
```
