```python
def pklopen(route):
    with open('{}'.format(route), 'rb') as f:
        df = pickle.load(f)
    return df
def pklsave(df, name):
    with open('{}.pkl'.format(name), 'wb') as f:
        pickle.dump(df, f)
```

```python
X, y_total, y59, y1014, y514 = Data.iloc[:,3:-4], Data.iloc[:,-4],Data.iloc[:,-3],Data.iloc[:,-2],Data.iloc[:,-1]
```

# 전연령 거주인구

```python
DT_model = DecisionTreeRegressor(max_depth=2)
RF_model = RandomForestRegressor(max_depth=6)
```

```python
DT_model.fit(X_train_total, y_train_total)
RF_model.fit(X_train_total, y_train_total)
print('DT_train: ', DT_model.score(X_train_total,y_train_total))
print('DT_test: ', DT_model.score(X_test_total,y_test_total))
print('RF_train: ', RF_model.score(X_train_total,y_train_total))
print('RF_test: ', RF_model.score(X_test_total,y_test_total))
```
