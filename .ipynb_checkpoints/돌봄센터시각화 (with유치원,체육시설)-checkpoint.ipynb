```python
map_osm = folium.Map(location=[37.5103, 126.982], zoom_start=15)
```

```python
for i in range(len(dolbom.index)):
    name = dolbom.loc[i, '시설명']
    lat = dolbom.loc[i, 'Latitude']
    lng = dolbom.loc[i, 'Longitude']
    
    marker = folium.Marker([lat,lng], popup=name,icon=folium.Icon(color='red'))
    circle = folium.CircleMarker([lat,lng],radius=80,color='#3186cc', fill_color='#3186cc')
    marker.add_to(map_osm)
    circle.add_to(map_osm)

# for n in dolbom.index:
#     folium.CircleMarker([dolbom['Latitude'][n], dolbom['Longitude'][n]],
#                         radius = dolbom['검거'][n]*10,
#                         color='#3186cc', fill_color='#3186cc').add_to(map)    

# map_osm.save('돌봄센터_분포.html')
```

```python
def calc_offsets(d, lat):
    return (
        abs(360*math.asin(math.sin(d/6371/2/1000)/math.cos(lat*math.pi/180))/math.pi),
        180*d/6371/1000/math.pi
    )
```

```python
def radian(degree):
    return math.acos(-1)/180*degree

def coordinate_after_rotation(c, degree, offsets):
    return (
        c[0]+math.cos(radian(degree))*offsets[0],
        c[1]+math.sin(radian(degree))*offsets[1]
    )
```
