```python
# 전체 구별 고령자 현황 그래프
def plot_region(df, x_axis, rot_x=True, show_xlabel=True):
    plt.figure(figsize=(18, 6))

    pal = sns.color_palette("Blues_r", df.shape[0])
    rank = df['65세이상_인구_수'].argsort().argsort()
    sns.barplot(x=x_axis, y='65세이상_인구_수', data=df, palette=np.array(pal[::-1])[rank])
    plt.axhline(0.075, color='red', alpha=0.5, label='서울시 동별 고령자 인구 수')
    for idx, score in enumerate(df['65세이상_인구_수']):
        plt.text(idx-0.4, score+0.01, score, size=10, weight='bold')
    plt.title(f'서울시 {x_axis}별 65세 이상 전체 인구', size=25)
    plt.legend(fontsize=15)
    plt.ylabel(f'65세 이상 {x_axis} 전체 인구 수', size=15)
    if rot_x:
        plt.xticks(rotation=45, size=15, ha='right')
    if show_xlabel:
        plt.xlabel(x_axis, size=15)
    else:
        plt.xlabel(None)

    plt.show()    
```

```python
pop_dong = elderly_20[(elderly_20['동'] != '합계')&(elderly_20['동'] != '소계')].reset_index(drop=True)
tmp_list = []
for dong in pop_dong['동'].unique():
    tmp_df = pop_dong[pop_dong['동']==dong]
    tmp_list.append(int(float(tmp_df['65세이상 인구'].values[0].replace(',',''))))

elderly_pop_dong = pd.DataFrame({'자치동명':pop_dong['동'].unique(), '65세이상_인구_수':tmp_list})\
                        .sort_values('65세이상_인구_수', ascending=False).reset_index(drop=True)

plot_region(df=elderly_pop_dong, x_axis='자치동명')
```

```python
# 동별 고령자 현황 그래프
def plot_dong_region(gu, x_axis, rot_x=True, show_xlabel=True):
    
    pop_dong = elderly_20[(elderly_20['자치구'] == f'{gu}') & (elderly_20['동'] != '합계')&(elderly_20['동'] != '소계')].reset_index(drop=True)
    tmp_list = []
    for dong in pop_dong['동'].unique():
        tmp_df = pop_dong[pop_dong['동']==dong]
        tmp_list.append(int(float(tmp_df['65세이상 인구'].values[0].replace(',',''))))

    elderly_pop_dong = pd.DataFrame({'자치동명':pop_dong['동'].unique(), '65세이상_인구_수':tmp_list})\
                        .sort_values('65세이상_인구_수', ascending=False).reset_index(drop=True)
    
    plt.figure(figsize=(18, 6))

    pal = sns.color_palette("Greens_r", elderly_pop_dong.shape[0])
    rank = elderly_pop_dong['65세이상_인구_수'].argsort().argsort()
    sns.barplot(x=x_axis, y='65세이상_인구_수', data=elderly_pop_dong, palette=np.array(pal[::-1])[rank])
    plt.axhline(0.075, color='red', alpha=0.5, label=f'서울시 {gu} 동별 고령자 인구 수')
    for idx, score in enumerate(elderly_pop_dong['65세이상_인구_수']):
        plt.text(idx-0.4, score+0.01, score, size=10, weight='bold')
    plt.title(f'서울시 {gu} {x_axis}별 65세 이상 전체 인구', size=25)
    plt.legend(fontsize=15)
    plt.ylabel(f'65세 이상 {x_axis} 전체 인구 수', size=15)
    if rot_x:
        plt.xticks(rotation=45, size=15, ha='right')
    if show_xlabel:
        plt.xlabel(x_axis, size=15)
    else:
        plt.xlabel(None)

    plt.show()
```

```python
plot_dong_region('송파구','자치동명')
```
